```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Sample data for demonstration: replace these with your actual model results
models = ['RELU', 'TanH', 'Leaky RELU', 'Sigmoid']

# Replace with the metrics from your models
accuracy = [0.86, 0.76, 0.93, 0.76]
precision = [0.88, 0.91, 0.94, 0.94]
recall = [0.86, 0.76, 0.93, 0.76]
f1 = [0.86, 0.78, 0.93, 0.81]

# Create a DataFrame for better plotting
import pandas as pd

metrics_df = pd.DataFrame({
    'Model': models,
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1
})

# Melt the DataFrame for easier plotting with Seaborn
metrics_melted = metrics_df.melt(id_vars='Model', var_name='Metric', value_name='Score')

# Set the plot size
plt.figure(figsize=(10, 6))

# Create a grouped bar plot using Seaborn
sns.barplot(x='Model', y='Score', hue='Metric', data=metrics_melted)

# Set the plot labels and title
plt.title('Model Comparison: Accuracy, Precision, Recall, F1-Score')
plt.ylabel('Score')
plt.ylim(0, 1)  # Set y-axis range between 0 and 1 for better comparison
plt.xticks(rotation=45)  # Rotate model names for readability
plt.legend(loc='lower right')

# Show the plot
plt.tight_layout()
plt.show()
```
